```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]

#plt.plot([],[],color='m', label='Sleeping', linewidth=5)
#plt.plot([],[],color='c', label='Eating', linewidth=5)
#plt.plot([],[],color='r', label='Working', linewidth=5)
#plt.plot([],[],color='k', label='Playing', linewidth=5)  

plt.stackplot(days, sleeping,eating,working,playing,colors=['m','c','r','k']) 
#anywhere where there is more than just a line, with things like fills or stackplots like this, we cannot label the specific parts inherently

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        #shadow= True,
        #explode=(0,0,0,0.1), which one is empasised 
        autopct='%1.1f%%') #fill in percentage

plt.title('Interesting Graph\nCheck it out')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

fig = plt.figure()

def create_plots():
    xs = []
    ys = []

    for i in range(10):
        x = i
        y = random.randrange(10)

        xs.append(x)
        ys.append(y)
    return xs, ys
```

```python
ax1 = fig.add_subplot(221) #221 means 2 tall, 2 wide, plot number 1. 212 is a 2 tall, 1 wide, plot number 1.
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)
x,y=create_plots()
ax1.plot(x,y)
ax2.plot(x,y)
ax3.plot(x,y)
plt.show()
```
